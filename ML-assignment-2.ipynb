```python
import pandas as pd
from datetime import datetime

customers = pd.read_csv("C:/Users/Shabd/Downloads/AWCustomers.csv")

customers['BirthDate'] = pd.to_datetime(customers['BirthDate'])
customers['Age'] = customers['BirthDate'].apply(lambda x: datetime.now().year - x.year)

selected = customers[['Age', 'YearlyIncome', 'TotalChildren', 'NumberChildrenAtHome', 'Education', 'Occupation', 'HomeOwnerFlag', 'NumberCarsOwned', 'Gender', 'MaritalStatus']]

df = selected.copy()
df.head()

types = {
    'Age': 'Continuous, Ratio',
    'YearlyIncome': 'Continuous, Ratio',
    'TotalChildren': 'Discrete, Ratio',
    'NumberChildrenAtHome': 'Discrete, Ratio',
    'Education': 'Discrete, Ordinal',
    'Occupation': 'Discrete, Nominal',
    'HomeOwnerFlag': 'Discrete, Nominal',
    'NumberCarsOwned': 'Discrete, Ratio',
    'Gender': 'Discrete, Nominal',
    'MaritalStatus': 'Discrete, Nominal'
}

types
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, KBinsDiscretizer, OneHotEncoder

df = df.fillna(df.mode().iloc[0])

scaler = MinMaxScaler()
df[['Age', 'YearlyIncome']] = scaler.fit_transform(df[['Age', 'YearlyIncome']])

kbins = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='quantile')
df['Income_Bin'] = kbins.fit_transform(df[['YearlyIncome']])

standard = StandardScaler()
df[['Age_std', 'YearlyIncome_std']] = standard.fit_transform(df[['Age', 'YearlyIncome']])

encoder = OneHotEncoder(drop='first', sparse=False)
encoded = encoder.fit_transform(df[['Education','Occupation','Gender','MaritalStatus']])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['Education','Occupation','Gender','MaritalStatus']))

final_df = pd.concat([df.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)

final_df.head()
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

numeric_df = final_df.select_dtypes(include=[np.number])

x = numeric_df.iloc[0].values
y = numeric_df.iloc[1].values

smc = np.sum(x == y) / len(x)

jaccard = np.sum((x == 1) & (y == 1)) / np.sum((x == 1) | (y == 1))

cosine = cosine_similarity([x], [y])[0][0]

smc, jaccard, cosine
```
